```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("local[*]") \
    .appName("lab-spark") \
    .getOrCreate()

spark
```

```python
!ls -R /content/drive/MyDrive
```

```python
df = spark.read.csv(
    "/content/drive/MyDrive/datasets/datasets/airlines.csv",
    header=True,
    inferSchema=True
)

df.show(5)

```

```python
df.printSchema()
```

```python
df.select("airline", "rating", "recommended").show(10)
```

```python
df_clean = df.filter(df.rating > 0)
df_clean.show(5)
```

```python
df.groupBy("airline").avg("rating").show()
```

```python
df.groupBy("recommended").count().show()
```

```python
df.write.mode("overwrite").parquet(
    "/content/drive/MyDrive/datasets/output/airlines_parquet"
)
```
